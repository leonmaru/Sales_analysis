```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path = (r"C:\Users\admin\OneDrive\Desktop\Panda sales analysis")
extension = 'csv'
files = [file for file in os.listdir(path)if file.endswith(extension)]
dfs = []
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    dfs.append(df)
    df = pd.concat(dfs,ignore_index = True)

df.head()
```

```python
df.shape
```

```python
# add specific column for month

df['Month'] = df['Order Date'].str[0:2]

df.head()
```

```python
# drop rows with NAN
nan_df = df[df.isna().any(axis = 1)]

df = df.dropna(how = 'all')

nan_df.head()
```

```python
# drop rows with 'Or'

df = df[df['Order Date'].str[0:2]!='Or']
df['Month'] = df['Month'].astype('int32')
```

```python
df.head()
```

```python
# add sales column
# make it numeric first
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])

df['Price Each'] = pd.to_numeric(df['Price Each'])

df['Sales'] = df['Quantity Ordered']*df['Price Each']

df.head()
```

```python
results = df.groupby('Month').sum()
results
```

```python
# plot a bar graph on Sales aganist Months

month = range(1,13)
plt.bar(month, results['Sales'])
plt.xticks(month)
plt.ylabel('sales in USD($)')
plt.xlabel('month number')

plt.show()
```

```python
# add city column
# lets use .apply()

def get_city(Address):
    return Address.split(',')[1]

def get_state(Address):
    return Address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x:get_city(x) + '('+ get_state(x)+')')

df.head()
```

```python
# what city had the highest number of sales?

results2 = df.groupby('City').sum()
results2
```

```python
# plot a bar graph on sales aganist cities

cities = [city for city, df in df.groupby('City')]
plt.bar(cities, results2['Sales'])
plt.xticks(cities,rotation = 'vertical', size=8)
plt.ylabel('Sales in USD($)')
plt.xlabel('City name')
plt.show()
```

```python
# what time should we display advertisement to maximise buying of products?

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Hour'] = df['Order Date'].dt.hour

hours = [hour for hour, df in df.groupby('Hour')]
plt.plot(hours, df.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()

```

```python
# count pairs of products occuring together
from itertools import combinations
from collections import counter

 count = counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```
